```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
import scipy.stats as sts
%matplotlib inline
```

```python
N = 1000
K = 10000
size = N*K
theta = 0;
stand_norm = sts.norm(theta, 1)
X = np.array(stand_norm.rvs(size))
X = X.reshape(N,K)
```

```python
est1 = np.array([X[:n].mean(axis=0) for n in range (1,N)]) # likelihood
```

По второй оценке

```python
est2 = np.array([(X[:n].min(axis=0) + X[:n].max(axis=0))/2 for n in range(1,N)]) # (min + max)/2
```

```python
sigma1 = np.array([est1[:n].var() for n in range (1, N)])
```

```python
sigma2 = np.array([est2[:n].var() for n in range (1,N)])
```

Теперь рисуем sigma1& sigma2

```python
plt.figure(figsize=(20,10))
plt.plot(np.array(range(1,N)), sigma1, color='green', label='sigma1')
plt.plot(np.array(range(1,N)), sigma2, color='red', label='sigma2')
plt.legend(loc='center left', bbox_to_anchor=(0.7, 0.7), fontsize = 30)
plt.ylim(0,0.3)
plt.show()
```
