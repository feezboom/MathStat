```python
import scipy.stats as sps
import numpy as np(0,1
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = sps.norm.rvs(size = 100, loc = 0, scale = 1)
```

## В модели N($\theta$, 1) найти оценку theta по макс правдоподобию

```python
est = np.array([np.array(X[:i]).mean() for i in range(1,101)])
```

# Теперь байесовские оценки

```python
def make_est(a,sigma_squared) :
    return np.array( [ (X[:n].sum() + (a/sigma_squared))/(n+(1/sigma_squared)) for n in range (1,101)] )
```

```python
est1 = make_est(0,1)
est2 = make_est(0,100)
est3 = make_est(10,1)
est4 = make_est(10,100)
```

```python
x = np.linspace(1,101,100)
plt.figure(figsize=(20,10))
plt.xlim(0, 100)
plt.ylim(0, 0.57)
plt.plot(x, abs(est), color = "red", label = 'likelihood')
plt.plot(x, abs(est1), color = "green", label = '(0,1)')
plt.plot(x, abs(est2), color = "magenta", label = '(0,100)')
plt.plot(x, abs(est3), color = "blue", label = '(10,1)')
plt.plot(x, abs(est3), color = "yellow", label = '(10,100)')
plt.legend(fontsize=30)
plt.show()
```

### Для параметров (10,100) оценка плохая, их лучше не использовать как параметры априорного распределения, а вот, к примеру, розовый график - самое то.

## В модели $N(0, \theta)$ найти оценку theta по макс правдоподобию

```python
est = np.array([np.array(X[:i]).var() for i in range(1,101)])
```

# Теперь Байесовские оценки

```python
def make_est(alpha, beta) :
    return np.array( [ (beta + ((X[:n]**2).sum()/2))/(alpha + (n/2.) - 1) for n in range (1,101)] )
```

```python
est1 = make_est(1,1)
est2 = make_est(1,100)
est3 = make_est(10,1)
est4 = make_est(10,100)
```

```python
x = np.linspace(1,101,100)
plt.figure(figsize=(20,10))
plt.xlim(0,100)
plt.ylim(0,50)
plt.plot(x, abs(est), color = "red", label = 'likelihood')
plt.plot(x, abs(est1), color = "green", label = '(1,1)')
plt.plot(x, abs(est2), color = "magenta", label = '(1,100)')
plt.plot(x, abs(est3), color = "blue", label = '(10,1)')
plt.plot(x, abs(est4), color = "yellow", label = '(10,100)')
plt.legend(fontsize=30)
plt.show()
```
