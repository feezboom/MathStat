```python
#   
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
%matplotlib inline

def printStatistics(N, theta) :
    # .
    X = sts.uniform.rvs(size=N, loc=0, scale=theta)

    result1 = []
    result2 = []
    result3 = []
    result4 = []
    result5 = []

    #2EX
    #EX + X(n)/2
    #X(1) * (n+1)
    #X(1) + X(n)
    #(n+1)/n * X(i)

    ourMax = X[0];
    ourMin = X[0];

    for i in range(0, N) :
        Xn = X[:i + 1]

        ourMax = max(ourMax, X[i]);
        ourMin = min(ourMin, X[i]);
        ourMean = Xn.mean()

        result1.append(abs(2 * ourMean - theta))
        result2.append(abs(ourMean + ourMax / 2 - theta))
        result3.append(abs(ourMin * i - theta))
        result4.append(abs(ourMin + ourMax - theta))
        result5.append(abs(((i+2)/(i+1))*ourMax - theta))
        
    
    
    
    plt.figure(figsize=(20,10))
    plt.ylim(0, theta/30)
    plt.title('$\theta=$' + str(theta), fontsize = 100);
    plt.plot(np.arange(1, N + 1), result1, color = 'red', label='$2\overline{X}$');
    plt.plot(np.arange(1, N + 1), result2, color = 'green', label='$\overline{X} + X_{(n)}/2$')
    plt.plot(np.arange(1, N + 1), result3, color = 'magenta', label='$(n+1)X_{(1)}$')
    plt.plot(np.arange(1, N + 1), result4, color = 'black', label='$X_{(1)}+X_{(n)}$')
    plt.plot(np.arange(1, N + 1), result5, color = 'blue', label=r'$\frac{n+1}{n} X_{(n)}$')
    plt.legend(loc='center left', bbox_to_anchor=(0.7, 0.7), fontsize = 30)
    plt.ylabel('$|\\theta - \\theta^*(n)|$', fontsize = 50);
    plt.show()

    print('\n\n\n\n\n\n\n\n\n')
    
N = 10000;
theta = 1;
for i in range(5) :
    printStatistics(N, theta);
    theta *= 10;
```
